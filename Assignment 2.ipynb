```python
df.info()
```

1) CRIM: per capita crime rate by town 
2) ZN: proportion of residential land zoned for lots over 25,000 sq.ft. 
3) INDUS: proportion of non-retail business acres per town 
4) CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise) 
5) NOX: nitric oxides concentration (parts per 10 million) 
6) RM: average number of rooms per dwelling 
7) AGE: proportion of owner-occupied units built prior to 1940 
8) DIS: weighted distances to five Boston employment centres 
9) RAD: index of accessibility to radial highways 
10) TAX: full-value property-tax rate per $10,000 
11) PTRATIO: pupil-teacher ratio by town 
12) B: The result of the equation B=1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
13) LSTAT: % lower status of the population

Output variable: 1) MEDV: Median value of owner-occupied homes in $1000's

```python
#Splitting of the dataset into X(input) and y(output)
X = df.iloc[:, :-1]
y = df['medv']  
```

```python
X
```

```python
y
```

```python
#This boxplot is giving us the way the distribution of values is.
sns.boxplot(data=X)
plt.show()
```

```python
##Splitting of entire dataset into train and test datasets with 70% as train dataset and 30% as test dataset and
##random state as 42 ensuring that the dataset is shuffled
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
# Standard scaler is used to standarize features.
# fit(): Computes the mean and standard deviation for scaling and transform(): Applies the transformation using those computed values.
#X_test must be transformed using the mean and standard deviation learned from X_train.
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(X_train)
X_train = pd.DataFrame(scaled_data, columns=X_train.columns)
scaled_data=scaler.transform(X_test)
X_test = pd.DataFrame(scaled_data, columns=X_test.columns)
```

```python
sns.boxplot(data=X_train)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=X_test)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# This model assumes a linear relationship between independent variables (X_train) and the target variable (y_train).
# The .fit() function trains the linear regression model using X_train (features) and y_train (target).
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
#prediction on the test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
reg_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': reg_pred})
reg_diff.head()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# This is used to measure how well the regression model fits the data using metrics 

mae= mean_absolute_error(y_test,reg_pred)
mse= mean_squared_error(y_test,reg_pred)
r2= r2_score(y_test,reg_pred)
rmse=np.sqrt(mean_squared_error(y_test,reg_pred))
print("Mean absolute error:",mae)
print("Mean squared error:",mse)
print("R squared error:",r2)
print("Root mean squared error:",rmse)
```
