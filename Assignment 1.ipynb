```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
df
```

```python
df.info()
```

```python
sns.histplot(data=df["Age"])
plt.show()
```

```python
# as the data is skewwd we will replace the null values in age with median 
print(df["Age"].median())
print(df["Age"].mean())
```

```python
df["Age"].fillna(df["Age"].median(),inplace=True)
```

```python
df["Age"].isnull().sum()
```

```python
# now as cabin and embarked are categorical columns so we will replace it with mode
print(df["Cabin"].mode())
print(df["Embarked"].mode())
```

```python
df["Cabin"]=df["Cabin"].fillna("Unknown")
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df=pd.get_dummies(df,columns=['Sex','Embarked'],drop_first=True)
df
```

```python
df["Sex_male"]=df["Sex_male"].astype(int)
df["Embarked_Q"]=df["Embarked_Q"].astype(int)
df["Embarked_S"]=df["Embarked_S"].astype(int)
```

```python
df
```

```python
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=df["Age"])
```

```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1-1.5*IQR
upper_bound = Q3-1.5*IQR

df_cleaned = df[(df["Age"]>=lower_bound) & (df["Age"]<=upper_bound)]
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
Q1_fare = df["Fare"].quantile(0.25)
Q3_fare = df["Fare"].quantile(0.75)

IQR_fare = Q3_fare - Q1_fare

lower_bound_fare = Q1_fare - 1.5 * IQR_fare
upper_bound_fare = Q3_fare + 1.5 * IQR_fare

df_cleaned = df_cleaned[(df_cleaned["Fare"] >= lower_bound_fare) & (df_cleaned["Fare"] <= upper_bound_fare)]
```

```python
sns.boxplot(data=df_cleaned)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df_cleaned[["Age","Fare"]])
df_scaled = pd.DataFrame(df_scaled, columns=['Age', 'Fare'])
```
